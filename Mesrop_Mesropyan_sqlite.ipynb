```python
import sqlite3
import pandas as pd


conn = sqlite3.connect('tips.db')

cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
pd.read_sql_query('''SELECT * FROM Observation WHERE total_bill>50''', conn)
```

```python
pd.read_sql_query('''SELECT AVG(tip) AS average_tip_given FROM Observation''', conn)
```

```python
pd.read_sql_query('''SELECT MAX(total_bill) AS maximum_total_bill FROM Observation''', conn)
```

```python
pd.read_sql_query('''
SELECT day, SUM(size) AS people_served
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day
''', conn)
```

```python
pd.read_sql_query('''
SELECT size, AVG(total_bill) AS avg_total_bill
FROM Observation
GROUP BY size
''', conn)
```

```python
pd.read_sql_query('''
SELECT Smoker.smoker, SUM(Observation.tip) AS total_tip_amount
FROM Observation
JOIN Smoker ON Observation.smoker_id = Smoker.smoker_id
GROUP BY Smoker.smoker
''', conn)
```

```python
pd.read_sql_query('''
SELECT day, (tip / total_bill) * 100 AS avg_tip_percentage
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day
''', conn)
```

# **<font color='red'> For average calculation you should either divide the sum of tip to the sum of total bill or use avg aggregate function on the division of tip to total bill. Minus 1 point </font>**

```python
pd.read_sql_query('''
SELECT Observation.total_bill, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
''', conn)
```

```python
pd.read_sql_query('''
SELECT Sex.sex, COUNT(*) AS total_number_of_observations
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex
''', conn)
```

```python
pd.read_sql_query('''
SELECT Observation.tip, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner'
''', conn)
```

```python
pd.read_sql_query('''
SELECT Sex.sex, MAX(Observation.total_bill) AS largest_total_bill
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex
''', conn)
```

```python
pd.read_sql_query('''
SELECT AVG(tip) AS average_tip
FROM Observation
WHERE size >4
''', conn)
```

```python
pd.read_sql_query('''
SELECT Observation.* , day, time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day IN ('Fri', 'Sat')
AND Time.time IN ('Lunch', 'Dinner')
''', conn)
```

```python
pd.read_sql_query('''
SELECT *
FROM Observation
WHERE (tip / total_bill) * 100 > 20
''', conn)
```

```python
pd.read_sql_query('''
SELECT *
FROM Observation
WHERE tip = (
    SELECT MAX(tip)
    FROM Observation
    WHERE tip < (SELECT MAX(tip) FROM Observation)
)
''', conn)
```
