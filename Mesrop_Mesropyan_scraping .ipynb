```python
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
resp = requests.get(url, headers=headers)
print(resp.status_code)
```

```python
soup = BeautifulSoup(resp.content, "html.parser")
import pandas as pd
movies = []
count =25
```

```python
movie_items = soup.find_all("li", class_="ipc-metadata-list-summary-item")
```

```python
for item in movie_items:
    title = item.find("h3", class_="ipc-title__text").get_text()

    rank = count
    count -=1
    metadata = item.find_all("span", class_="cli-title-metadata-item")
    year = metadata[0].get_text()
    duration = metadata[1].get_text()
    age_rating = metadata[2].get_text()

    rating_group = item.find("span", class_="ipc-rating-star--rating")
    rating = rating_group.get_text()

    vote_group = item.find("span", class_="ipc-rating-star--voteCount")
    votes = vote_group.get_text()

    movies.append({
        "Rank": rank,
        "Title": title,
        "Year": year,
        "Duration": duration,
        "Age Rating": age_rating,
        "IMDb Rating": rating,
        "Votes": votes
    })
```

```python
df = pd.DataFrame(movies)
```
